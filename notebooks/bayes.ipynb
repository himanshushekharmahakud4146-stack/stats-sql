## What is Bayes' Theorem?
P(A|B) = P(B|A) * P(A) / P(B)

It answers: "Given that B happened, what is the probability of A?"

```python
# Real example: Email spam classifier

# Prior: 30% of all emails are spam
P_spam = 0.30
P_not_spam = 0.70

# Likelihood: word "FREE" appears in 80% of spam, 10% of non-spam
P_free_given_spam = 0.80
P_free_given_not_spam = 0.10

# Total probability of seeing "FREE"
P_free = (P_free_given_spam * P_spam) + (P_free_given_not_spam * P_not_spam)

# Bayes: P(spam | "FREE" in email)
P_spam_given_free = (P_free_given_spam * P_spam) / P_free

print(f"P(FREE appears) = {P_free:.2f}")
print(f"P(spam | FREE) = {P_spam_given_free:.2f}")
print(f"If email contains 'FREE', there is a {P_spam_given_free*100:.1f}% chance it is spam")
```

```python
def bayes_spam_classifier(word, p_word_spam, p_word_not_spam, p_spam=0.3):
    """
    Classify if email is spam based on one word using Bayes theorem.
    """
    p_not_spam = 1 - p_spam
    
    # Total probability of the word appearing
    p_word = (p_word_spam * p_spam) + (p_word_not_spam * p_not_spam)
    
    # Bayes formula
    p_spam_given_word = (p_word_spam * p_spam) / p_word
    
    return p_spam_given_word

# Test with different words
words = {
    "FREE":    (0.80, 0.10),
    "meeting": (0.05, 0.40),
    "winner":  (0.70, 0.05),
    "project": (0.10, 0.35),
}

print("Word         | Spam Probability")
print("-" * 35)
for word, (p_spam_w, p_not_spam_w) in words.items():
    prob = bayes_spam_classifier(word, p_spam_w, p_not_spam_w)
    print(f"{word:<12} | {prob*100:.1f}%")
```

## My Understanding of Bayes in 5 Lines

1. Bayes theorem updates our belief when we see new evidence.
2. P(A|B) means: probability of A, given that B already happened.
3. We start with a "prior" belief (like 30% emails are spam).
4. After seeing evidence (like the word FREE), we update to a "posterior".
5. This is powerful because machines can learn to classify using just probabilities.
