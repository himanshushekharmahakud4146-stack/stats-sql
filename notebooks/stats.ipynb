```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head())
```

```python
df.describe()
```

```python
col = 'sepal length (cm)'

print("=== MEAN ===")
print(df[col].mean())

print("\n=== MEDIAN ===")
print(df[col].median())

print("\n=== MODE ===")
print(df[col].mode()[0])

print("\n=== All columns MEAN ===")
print(df.mean())

print("\n=== All columns MEDIAN ===")
print(df.median())

print("\n=== Variance ===")
print(df[col].var())

print("\n=== Std Deviation ===")
print(df[col].std())

print("\n=== Min & Max ===")
print("Min:", df[col].min())
print("Max:", df[col].max())

print("\n=== Range ===")
print(df[col].max() - df[col].min())
```

```python
df['sepal length (cm)'].hist(bins=20, color='steelblue', edgecolor='black')
plt.title('Sepal Length Distribution')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```
