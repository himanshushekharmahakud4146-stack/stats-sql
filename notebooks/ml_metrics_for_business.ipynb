# ML Metrics for Business

## Objective

Explain machine learning evaluation metrics using real-world BFSI examples.

## Topics

- Confusion Matrix
- Accuracy
- Precision
- Recall
- F1 Score
- ROC-AUC

# Why Metrics Matter

A machine learning model can make predictions, but businesses need a way to measure how reliable those predictions are.

Metrics help organizations evaluate model performance and make informed decisions.

# Business Scenario 1: Fraud Detection

Goal:

Identify fraudulent transactions before financial loss occurs.

Challenge:

Fraud cases are rare, creating an imbalanced dataset.

Business Impact:

Missing a fraud transaction can lead to financial losses.

# Business Scenario 2: Loan Default Prediction

Goal:

Predict whether a customer will fail to repay a loan.

Business Impact:

Incorrect approvals increase financial risk.

Incorrect rejections reduce business opportunities.

# Confusion Matrix

TP = Correct Positive Prediction

TN = Correct Negative Prediction

FP = Incorrect Positive Prediction

FN = Incorrect Negative Prediction

# Accuracy

Formula:

Accuracy = (TP + TN) / (TP + TN + FP + FN)

Business Meaning:

Measures overall prediction correctness.

# Precision

Formula:

Precision = TP / (TP + FP)

Business Meaning:

When the model predicts fraud, how often is it correct?

# Recall

Formula:

Recall = TP / (TP + FN)

Business Meaning:

Out of all fraud transactions, how many did the model detect?

# F1 Score

Formula:

F1 = 2 × (Precision × Recall) / (Precision + Recall)

Business Meaning:

Balances Precision and Recall.

# ROC-AUC

Business Meaning:

Measures how well the model separates positive and negative classes.

Higher AUC indicates better model performance.

# When To Prefer Precision

Fraud Detection

A false fraud alert may inconvenience customers.

High Precision reduces false alarms.

# When To Prefer Recall

Fraud Detection

Missing a fraud transaction is costly.

High Recall helps detect more fraud cases.

# Key Business Takeaways

1. Accuracy alone is not enough.

2. Precision measures prediction quality.

3. Recall measures detection capability.

4. F1 Score balances Precision and Recall.

5. ROC-AUC evaluates overall classifier quality.

6. Business requirements determine which metric is most important.
