# SQL String Functions & Pattern Matching Analysis

## Objective

The objective of this notebook is to practice SQL string functions and pattern matching using PostgreSQL. These functions are commonly used in data analysis, reporting, and data cleaning.

---

# Dataset

Database: **flipkart_db**

Table: **products**

---

# 1. View All Products

### Explanation

Retrieve all records from the products table.

```sql
SELECT *
FROM products;
```

---

# 2. Find Unique Categories

### Explanation

DISTINCT removes duplicate values and returns only unique categories.

```sql
SELECT DISTINCT category
FROM products;
```

---

# 3. Convert Categories to Uppercase

### Explanation

UPPER() converts every character into uppercase.

```sql
SELECT DISTINCT
UPPER(category)
FROM products;
```

---

# 4. Convert Product Names to Lowercase

### Explanation

LOWER() converts text into lowercase.

```sql
SELECT
LOWER(name)
FROM products;
```

---

# 5. Count Characters in Product Names

### Explanation

LENGTH() returns the total number of characters.

```sql
SELECT
name,
LENGTH(name)
FROM products;
```

---

# 6. Search Using LIKE

### Products starting with W

```sql
SELECT *
FROM products
WHERE name LIKE 'W%';
```

---

### Products ending with r

```sql
SELECT *
FROM products
WHERE name LIKE '%r';
```

---

### Products containing "oo"

```sql
SELECT *
FROM products
WHERE name LIKE '%oo%';
```

---

### Products with second letter "a"

```sql
SELECT *
FROM products
WHERE name LIKE '_a%';
```

---

# 7. Rename Column Using Alias

### Explanation

Aliases improve readability by assigning temporary names.

```sql
SELECT
UPPER(category) AS category_upper
FROM products;
```

---

# 8. Combined Query

```sql
SELECT DISTINCT
UPPER(category) AS category_upper
FROM products
ORDER BY category_upper DESC;
```
