# Handling Imbalanced Datasets

## Objective

To understand the challenges of imbalanced datasets and explore common techniques used to improve machine learning performance on minority classes.

## Concepts Covered

- Imbalanced Data
- Oversampling
- Undersampling
- SMOTE (Synthetic Minority Oversampling Technique)
- Ensemble Methods
- Cost-Sensitive Learning
- Accuracy vs F1 Score

## Dataset

Credit Card Fraud Detection Dataset

## Goal

Demonstrate why high accuracy can be misleading and explore techniques for improving minority class prediction.

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)

from imblearn.over_sampling import SMOTE

import matplotlib.pyplot as plt
import seaborn as sns
```

# Imbalanced Dataset

An imbalanced dataset occurs when one class contains significantly more samples than another.

Example:

Normal Transactions = 99%

Fraud Transactions = 1%

This causes machine learning models to favor the majority class.

As a result, high accuracy may not indicate good performance.

# Oversampling

Oversampling increases the number of minority class samples.

Original:

Normal = 990
Fraud = 10

After Oversampling:

Normal = 990
Fraud = 990

```python
import matplotlib.pyplot as plt

classes = ["Normal","Fraud"]
counts = [990,990]

plt.scatter(classes, counts, s=200)

plt.title("Oversampling")

plt.ylabel("Count")

plt.show()
```

# Undersampling

Undersampling reduces the majority class size.

Original:

Normal = 990
Fraud = 10

After Undersampling:

Normal = 10
Fraud = 10

```python
classes = ["Normal","Fraud"]
counts = [10,10]

plt.scatter(classes, counts, s=200)

plt.title("Undersampling")

plt.ylabel("Count")

plt.show()
```

# SMOTE

SMOTE creates synthetic minority class samples instead of duplicating existing records.

Original:

Normal = 990
Fraud = 10

After SMOTE:

Normal = 990
Fraud = 990 (synthetic samples generated)

```python
classes = ["Normal","Fraud"]
counts = [990,990]

plt.scatter(classes, counts, s=200)

plt.title("SMOTE Balanced Dataset")

plt.ylabel("Count")

plt.show()
```

# Key Insight

The normal model has higher accuracy.

However, its F1 Score is very low because it fails to identify fraud transactions.

The balanced model has slightly lower accuracy but a much higher F1 Score.

For imbalanced datasets, F1 Score is often more informative than Accuracy.
