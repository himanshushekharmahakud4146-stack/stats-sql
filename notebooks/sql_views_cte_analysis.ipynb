# SQL Views & Introduction to CTEs (PostgreSQL)

## Objective

Learn how SQL Views simplify repeated queries and understand the purpose of Common Table Expressions (CTEs) for writing cleaner SQL.

---

# Topics Covered

## SQL Views

A View is a virtual table created from the result of a SQL query.

Views do not duplicate data. Instead, they store the query definition and return fresh results whenever they are queried.

---

## CREATE VIEW

Example

```sql
CREATE VIEW fitness_item AS
SELECT product_id,
       name,
       stock_quantity,
       price
FROM products
WHERE category = 'Fitness';
```

Querying the view:

```sql
SELECT *
FROM fitness_item;
```

---

## Another View

```sql
CREATE VIEW low_stocks AS
SELECT category,
       stock_quantity
FROM products
WHERE stock_quantity > 20;
```

Retrieve the data:

```sql
SELECT *
FROM low_stocks;
```

---

# Introduction to CTE

A Common Table Expression (CTE) is a temporary named result set created using the WITH clause.

Example

```sql
WITH expensive_products AS (
    SELECT *
    FROM products
    WHERE price > 1000
)
SELECT *
FROM expensive_products;
```

---

# Difference Between Views and CTEs

| View | CTE |
|------|------|
| Permanent database object | Temporary query |
| Can be reused many times | Exists only for one SQL statement |
| Stored inside database | Created using WITH clause |
| Best for frequently used queries | Best for simplifying complex queries |

# Learning Summary
