## 1. Basic SELECT & Filtering

```python
query = """
SELECT *
FROM products;
"""

df = pd.read_sql(query, engine)

df
```

```python
query = """
SELECT
    name,
    price
FROM products;
"""

pd.read_sql(query, engine)
```

```python
query = """
SELECT *
FROM products
WHERE category = 'Electronics';
"""

pd.read_sql(query, engine)
```

```python
query = """
SELECT *
FROM products
WHERE price > 500;
"""

pd.read_sql(query, engine)
```

```python
query = """
SELECT *
FROM products
WHERE price > 1000
AND category = 'Electronics';
"""

pd.read_sql(query, engine)
```

```python
query = """
SELECT *
FROM products
WHERE price BETWEEN 400 AND 1000;
"""

pd.read_sql(query, engine)
```

```python
query = """
SELECT *
FROM products
WHERE category IN
('Electronics',
 'Home & Kitchen',
 'Fitness');
"""

pd.read_sql(query, engine)
```

```python
query = """
SELECT *
FROM products
WHERE NOT category = 'Electronics';
"""

pd.read_sql(query, engine)
```
