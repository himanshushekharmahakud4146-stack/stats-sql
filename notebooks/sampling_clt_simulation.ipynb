```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
population = np.random.randint(1, 100, 10000)
```

```python
sample_sizes = [10, 30, 100]
```

```python
for n in sample_sizes:
    
    sample_means = []
    
    for i in range(1000):
        sample = np.random.choice(population,
                                  size=n,
                                  replace=True)
        
        sample_means.append(np.mean(sample))
    
    plt.figure(figsize=(8,4))
    
    plt.hist(sample_means,
             bins=30)
    
    plt.title(f"Sampling Distribution (n={n})")
    
    plt.xlabel("Sample Mean")
    
    plt.ylabel("Frequency")
    
    plt.show()
```

# Central Limit Theorem (CLT)

The Central Limit Theorem states that the distribution of sample means tends to become approximately normal as the sample size increases.

Even if the original population is not normally distributed, the sampling distribution of the mean approaches a normal distribution.

In this simulation, larger sample sizes (30 and 100) produced a more bell-shaped distribution than smaller sample sizes.
