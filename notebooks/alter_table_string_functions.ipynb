## String Functions

### UPPER()

Converts text into uppercase.

Example

```sql
SELECT UPPER(name)
FROM products;
```

---

### LOWER()

Converts text into lowercase.

```sql
SELECT LOWER(name)
FROM products;
```

---

### LENGTH()

Returns number of characters.

```sql
SELECT LENGTH(sku_code)
FROM products;
```

---

### SUBSTRING()

Extracts part of a string.

```sql
SELECT SUBSTRING('I am learning SQL',6,12);
```

---

### LEFT()

Returns left characters.

```sql
SELECT LEFT(sku_code,2)
FROM products;
```

---

### RIGHT()

Returns right characters.

```sql
SELECT RIGHT(sku_code,3)
FROM products;
```

---

### CONCAT()

Joins multiple strings.

```sql
SELECT CONCAT(name,' | ',sku_code,' | ',category)
FROM products;
```

---

### TRIM()

Removes leading and trailing spaces.

```sql
SELECT TRIM('    hello SQL    ');
```

---

### REPLACE()

Replaces matching text.

```sql
SELECT REPLACE(sku_code,LEFT(sku_code,2),'FF')
FROM products;
```
