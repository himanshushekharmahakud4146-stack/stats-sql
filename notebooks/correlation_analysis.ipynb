# Correlation Analysis Using Titanic Dataset

## Objective
To explore relationships between variables using correlation analysis.

## Tools Used
- Python
- Pandas
- Seaborn
- Matplotlib

## Dataset
Titanic Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("titanic")

df.head()
```

```python
numeric_df = df.select_dtypes(include=['number'])

corr_matrix = numeric_df.corr()

corr_matrix
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm"
)

plt.title("Titanic Correlation Heatmap")

plt.show()
```

# Business Insights

1. Passenger class has a strong relationship with fare.

2. Higher fares are associated with better survival rates.

3. Passenger class and survival show meaningful correlation.

4. Age has relatively weak correlation with survival.

5. Fare can be an important predictor in survival-related models.

# Summary

Correlation analysis helps identify relationships between variables.

The heatmap provides a visual representation of correlation strengths and can assist in feature selection and exploratory data analysis.
