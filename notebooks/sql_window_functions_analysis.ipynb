# SQL Window Functions Analysis

## Objective

Learn advanced SQL window functions to perform ranking and analytical operations while preserving all rows in the result set.

---

# Concepts Covered

- OVER()
- PARTITION BY
- ROW_NUMBER()
- RANK()
- DENSE_RANK()
- Window Aggregates
- MAX() OVER()
- Ranking inside groups

---

# Dataset

Table: customers

Columns

- customer_id
- customer_name
- region
- revenue

---

# Queries Practiced

## GROUP BY

Find maximum revenue by region.

```sql
SELECT region,
       MAX(revenue)
FROM customers
GROUP BY region;
```

---

## MAX() OVER()

Display each customer together with the maximum revenue of their region.

---

## ROW_NUMBER()

Assign a unique row number inside every region.

---

## RANK()

Rank customers by revenue.

Customers with equal revenue receive the same rank.

---

## DENSE_RANK()

Rank customers without skipping numbers after ties.

---

## PARTITION BY

Restart ranking independently inside every region.

---

## Business Use Cases

- Top customers per region
- Employee salary ranking
- Leaderboards
- Sales analytics
- Department-wise reporting

---

# Key Learning

Unlike GROUP BY, window functions keep every row while adding analytical information such as rankings, totals, averages, and maximum values.

Window functions are widely used in Data Analytics, Business Intelligence, and SQL interviews.
