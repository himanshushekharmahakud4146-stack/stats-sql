# SQL Time Series Analysis

## Objective

Learn SQL window functions for time-series analysis using **LAG()**, **LEAD()**, and **Running Totals**.

---

# Concepts Covered

- OVER()
- ORDER BY inside window functions
- LAG()
- LEAD()
- SUM() OVER()
- Running Totals
- Month-over-Month (MoM) Revenue Change

---

# Dataset

Table: monthly_sales

Columns

- month_id
- month_name
- revenue

---

# Queries Practiced

## Display Monthly Sales

```sql
SELECT *
FROM monthly_sales;
```

---

## Previous Month Revenue using LAG()

```sql
SELECT
    month_name,
    revenue,
    LAG(revenue)
    OVER(ORDER BY month_id)
AS previous_month
FROM monthly_sales;
```

---

## Month-over-Month Revenue Change

```sql
SELECT
    month_name,
    revenue,
    revenue -
    LAG(revenue)
    OVER(ORDER BY month_id)
AS revenue_change
FROM monthly_sales;
```

---

## Next Month Revenue using LEAD()

```sql
SELECT
    month_name,
    revenue,
    LEAD(revenue)
    OVER(ORDER BY month_id)
AS next_month
FROM monthly_sales;
```

---

## Running Total

```sql
SELECT
    month_name,
    revenue,
    SUM(revenue)
    OVER(ORDER BY month_id)
AS running_total
FROM monthly_sales;
```

---

# Business Use Cases

- Sales trend analysis
- Monthly revenue comparison
- Forecasting future sales
- Dashboard reporting
- Financial analytics

---

# Key Learning

LAG() accesses the previous row.

LEAD() accesses the next row.

SUM() OVER() calculates cumulative totals without collapsing rows.
